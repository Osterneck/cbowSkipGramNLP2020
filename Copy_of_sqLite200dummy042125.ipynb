```python
!pip install faker

import sqlite3
from faker import Faker

# Setup
conn = sqlite3.connect("dummy_data.db")
cursor = conn.cursor()
fake = Faker()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    address TEXT,
    phone TEXT,
    email TEXT
)
""")

# Insert 200 dummy records
for _ in range(200):
    cursor.execute("""
    INSERT INTO users (name, address, phone, email)
    VALUES (?, ?, ?, ?)
    """, (fake.name(), fake.address(), fake.phone_number(), fake.email()))

# Save & close
conn.commit()
conn.close()
```
